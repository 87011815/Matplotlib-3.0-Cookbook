```python
import pandas as pd
import matplotlib.pyplot as plt

wine_quality = pd.read_csv('winequality.csv', delimiter=';')
iris = pd.read_csv('iris_dataset.csv', delimiter=',')
iris_mean = iris.mean()
iris_std = iris.std()

fig, axs = plt.subplots(2, 2, figsize=(8, 8))
axs[0, 0].hist(wine_quality['alcohol'])
axs[0, 0].set_title('Histogram')
axs[0, 0].set_xlabel('Alcohol Bins')
axs[0, 0].set_ylabel('Frequency')

axs[0, 1].plot(iris['sepal_length'], iris['sepal_width'])
axs[0, 1].set_title('Line')
axs[0, 1].set_xlabel('Sepal Length')
axs[0, 1].set_ylabel('Sepal Width')

axs[1, 0].scatter(iris['petal_length'], iris['petal_width'])
axs[1, 0].set_title('Scatter')
axs[1, 0].set_xlabel('Petal Length')
axs[1, 0].set_ylabel('Petal Width')

axs[1, 1].bar(['sepal_l','sepal_w', 'petal_l', 'petal_w'], iris_mean, yerr=iris_std)
#axs[1, 1].bar([0, 1, 2, 3], iris_mean, yerr=iris_std)
axs[1, 1].set_title('Bar')
axs[1, 1].set_xlabel('Category')
axs[1, 1].set_ylabel('Category Mean')

plt.suptitle('Subplots Demo')
plt.tight_layout(pad=3, w_pad=1.0, h_pad=1.0)
plt.show()
```

### Logarithmic scale

```python
import matplotlib.pyplot as plt

fig, (ax1, ax2, ax3) = plt.subplots(1, 3)

ax1.set_xlim(1e1, 1e3)
ax1.set_ylim(1e2, 1e4)
ax1.plot([0.1, 5, 10, 500], [0.01, 25, 100, 10000], "x-")
ax1.set_title("Linear Scale")

ax2.set_xscale("log")
ax2.set_yscale("log")
ax2.set_xlim(1e1, 1e3)
ax2.set_ylim(1e2, 1e3)
ax2.set_aspect(2)
ax2.set_title("adjustable = box")

ax3.set_xscale("log")
ax3.set_yscale("log")
ax3.set_adjustable("datalim")
ax3.plot([0.1, 5, 10, 500], [0.01, 25, 100, 10000], "o-")
ax3.set_xlim(1e-1, 1e3)
ax3.set_ylim(1e-2, 1e4)
ax3.set_aspect(1)
ax3.set_title("adjustable = datalim")

plt.tight_layout()

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

menMeans = [3, 10, 100, 500, 50]
menStd = [0.75, 2.5, 25, 125, 12.5]

fig, ax = plt.subplots()
ind = np.arange(len(menMeans))                # the x locations for the groups
width = 0.35                                  # the width of the bars
p1 = ax.bar(ind, menMeans, width, color='r', bottom=0, yerr=menStd)

womenMeans = [1000, 3, 30, 800, 1]
womenStd = [500, 0.75, 8, 200, 0.25]

p2 = ax.bar(ind + width, womenMeans, width, color='y', bottom=0, yerr=womenStd)
ax.set_title('Scores by category and gender')
ax.set_xticks(ind + width / 2)
ax.set_xticklabels(('C1', 'C2', 'C3', 'C4', 'C5'))

ax.set_yscale('log')
ax.legend((p1[0], p2[0]), ('Men', 'Women'))

plt.show()
```

### Using Units

```python
"""
============
Artist tests
============

Test unit support with each of the Matplotlib primitive artist types.

The axis handles unit conversions and the artists keep a pointer to their axis
parent. You must initialize the artists with the axis instance if you want to
use them with unit data, or else they will not know how to convert the units
to scalars.
"""

import matplotlib.lines as lines
import matplotlib.text as text
from basic_units import cm, inch
import matplotlib.pyplot as plt

fig, ax = plt.subplots(1,2)
# test a plain-ol-line

line = lines.Line2D([0*cm, 1.5*cm], [0*cm, 2.5*cm], lw=2, color='black', axes=ax[0])
t = text.Text(3*cm, 2.5*cm, 'text label', ha='left', va='bottom', axes=ax[0])

ax[0].add_line(line)
ax[0].add_artist(t)
ax[0].set_xlim(-1*cm, 10*cm)
ax[0].set_ylim(-1*cm, 10*cm)
ax[0].xaxis.set_units(cm)
ax[0].yaxis.set_units(cm)
ax[0].grid(True)

line = lines.Line2D([0*cm, 1.5*cm], [0*cm, 2.5*cm], lw=2, color='black', axes=ax[1])
t = text.Text(3*cm, 2.5*cm, 'text label', ha='left', va='bottom', axes=ax[1])

ax[1].add_artist(line)
ax[1].add_artist(t)
ax[1].set_xlim(-1*cm, 10*cm)
ax[1].set_ylim(-1*cm, 10*cm)
ax[1].xaxis.set_units(inch)
ax[1].yaxis.set_units(inch)
ax[1].grid(True)

plt.suptitle("Demonstration of Units Of Measurement")

plt.tight_layout(pad=3)

plt.show()
```

```python
import numpy as np
from basic_units import cm, inch
import matplotlib.pyplot as plt

cms = cm * np.arange(0, 21, 3)
bottom = 0 * cm
width = 0.8 * cm

fig, axs = plt.subplots(2, 2)

axs[0, 0].bar(cms, cms, bottom=bottom)
axs[0, 0].set_xticks(cm * np.arange(0, 21, 3))

axs[0, 1].bar(cms, cms, bottom=bottom, width=width, xunits=cm, yunits=inch)
axs[0, 1].set_xticks(cm * np.arange(0, 21, 3))

axs[1, 0].bar(cms, cms, bottom=bottom, width=width, xunits=inch, yunits=cm)
axs[1, 0].set_xticks(cm * np.arange(0, 21, 3))
axs[1, 0].set_xlim(2, 6)  # scalars are interpreted in current units

axs[1, 1].bar(cms, cms, bottom=bottom, width=width, xunits=inch, yunits=inch)
axs[1, 1].set_xticks(cm * np.arange(0, 21, 3))
axs[1, 1].set_xlim(2 * cm, 6 * cm)  # cm are converted to inches

fig.tight_layout()
plt.show()
```
